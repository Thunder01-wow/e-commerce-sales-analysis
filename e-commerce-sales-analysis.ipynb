# E-Commerce Sales Analysis

## Project Overview
This project analyzes an e-commerce sales dataset to understand sales trends,
profitability, and customer behavior across different categories and segments.

## Objectives
- Analyze overall sales and profit performance
- Identify top-performing categories and sub-categories
- Study monthly sales trends
- Compare sales and profit across customer segments

```python
import pandas as pd

import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio
pio.renderers.default = "svg"
import plotly.colors as colors
pio.templates.default = "plotly_white"
```

## Dataset Description
The dataset contains transaction-level sales data including:
- Order Date: Date of purchase
- Category / Sub-Category: Product classification
- Segment: Customer segment
- Sales: Revenue generated
- Profit: Net profit

## Data Loading

```python
data=pd.read_csv("sample.csv",encoding="latin-1")
```

## Data Cleaning and Quality Checks

```python
data["Order Date"]=pd.to_datetime(data["Order Date"])
data["Ship Date"]=pd.to_datetime(data["Ship Date"])
```

```python
data.info()
```

```python
data.duplicated().sum()
```

```python
data.head()
```

```python
data["Order Month"]=data["Order Date"].dt.month
data["Order Year"]=data["Order Date"].dt.year
data["Order Day Of Week "]=data["Order Date"].dt.dayofweek
```

```python
data.head()
```

## Data Cleaning
- Converted date columns to datetime format
- Checked for missing and duplicate values
- No major data quality issues were found

## Key Performance Indicators (KPIs)

```python
total_sales = data['Sales'].sum()
total_profit = data['Profit'].sum()
profit_margin = (total_profit / total_sales) * 100

total_sales, total_profit, profit_margin
```

### KPI Summary
- Total Sales: $ 2297200.86
- Total Profit: $ 286397.02
- Overall Profit Margin: 12.46%


# Monthly sales 

```python
sales_by_month=data.groupby("Order Month")["Sales"].sum().reset_index()

sales_by_month
```

```python
fig = px.line(sales_by_month,
              x="Order Month",
              y="Sales",
              title="Monthly Sales Analysis")
fig.show()
```

### Insights
- Sales peak during the later months of the year
- Indicates seasonal demand patterns
- Useful for inventory and marketing planning

```python
data.head()
```

# Sales by Category

```python
sales_by_category=data.groupby("Category")["Sales"].sum().reset_index()

sales_by_category
```

```python
fig= px.pie(sales_by_category,
           values="Sales",
           names="Category",
           hole=0.2,
           color_discrete_sequence=px.colors.qualitative.Pastel)

fig.update_traces(textposition ="inside" , textinfo ="percent + label")
fig.update_layout(title_text="Sales Analysis by Category")

fig.show()
```

### Insights
- Technology category generates the highest sales
- Furniture has comparatively lower profit margins
- Office Supplies show stable performance

# Profit by Category and Sub-Category

# Sales and Profit by Customer Segment

```python
data.head()
```

```python
sales_profit_by_segment = data.groupby("Segment").agg({"Sales" : "sum" , "Profit" : "sum"}).reset_index()
sales_profit_by_segment
```

```python
fig = px.bar(
    sales_profit_by_segment,
    x="Segment",
    y=["Sales", "Profit"],  
    barmode="group",         
    title="Sales and Profit Analysis by Customer Segment",
    color_discrete_sequence=px.colors.qualitative.Pastel
)

fig.update_layout(
    xaxis_title="Customer Segment",
    yaxis_title="Amount"
)

fig.show()
```

# Sales to Profit Ratio

```python
sales_profit_by_segment = data.groupby("Segment").agg({"Sales" : "sum" , "Profit" : "sum"}).reset_index()
sales_profit_by_segment["Sales_to_Profit_Ratio"] = sales_profit_by_segment["Sales"] / sales_profit_by_segment["Profit"]
sales_profit_by_segment[["Segment" , "Sales_to_Profit_Ratio"]]
```
